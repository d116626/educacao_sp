```python
import pandas as pd
import numpy as np
import os

pd.options.display.max_columns = 999
pd.options.display.max_rows = 999

pd.options.display.max_colwidth = 100

import geopandas as gpd

```

```python
maps = gpd.read_file('../maps/sp_municipios.json')
maps = maps[['geocodigo','nome_uf','nome_mesorregiao','nome_microrregiao','nome_municipio']]
```

```python
df = pd.read_csv('escolas_sp.csv')
df['matriculas']=1
matriculas = df[['CO_MUNICIPIO','matriculas']].groupby(by=['CO_MUNICIPIO'], as_index=False).sum()
```

```python
matriculas.head()
```

```python
df_final = pd.merge(matriculas,maps, left_on='CO_MUNICIPIO', right_on='geocodigo')
```
